# 4. Menentukan Nilai Prior

Perhitungan nilai prior dalam Naive Bayes dilakukan dengan menghitung proporsi atau probabilitas kemunculan setiap kelas dalam dataset. 

**Nilai prior = jumlah frekuensi kemunculan kelas / dengan total jumlah data**

Berikut adalah langkah-langkah :

1. Menghitung Jumlah Data: Hitung total jumlah data dalam dataset.
2. Menghitung Jumlah Kemunculan Setiap Kelas: Hitung jumlah kemunculan setiap kelas (ClassAttrition) dalam dataset.
3. Menghitung Nilai Prior untuk Setiap Kelas: Hitung nilai prior untuk setiap kelas dengan membagi jumlah kemunculan kelas tersebut dengan total jumlah data.

```python
# Target Class
target_class = 'ClassAttrition'

# Tampilkan Value dari target class
target_class_values = df[target_class].unique()
print(f"Nilai unik dari target class: {target_class_values}\n")

# Menentukan Nilai Prior

prior = {}

# Jumlah total data
n_T = len(df)
prior['n(T)'] = n_T
print(f"Jumlah total data (n(T)): {n_T}")

# Jumlah data untuk setiap nilai target class
for idx, target in enumerate(target_class_values):
    n_Ai = len(df[df[target_class] == target])
    prior[f'n(A{idx + 1})'] = n_Ai
    print(f"Jumlah data dengan target class {target} (n(A{idx + 1})): {n_Ai}")

# Menghitung prior probability
for idx, target in enumerate(target_class_values):
    n_Ai = len(df[df[target_class] == target])
    p_Ai = n_Ai / n_T
    prior[f'P(A{idx + 1})'] = p_Ai
    print(f"Prior probability untuk target class {target} (P(A{idx + 1})): {p_Ai}")
```

# 5. Tabel Kontigensi

Tabel kontingensi digunakan untuk menghitung nilai-nilai yang diperlukan untuk mengestimasi parameter model

```python
# Kontigensi dari setiap atribut

attributes = df.columns.tolist()

for attribute in attributes:
    if attribute != target_class:
        print("=" * 50)
        print(f"Kontigensi dari atribut \"{attribute}\"")
        print("-" * 50)
        for target in target_class_values:
            print(f"{target_class} \"{target}\"")
            attribute_class_value = df[attribute].unique().tolist()
            total = 0
            print(f"{attribute} :")
            for attribute_value in attribute_class_value:
                n_Ai = len(df[(df[attribute] == attribute_value) & (df[target_class] == target)])
                print(f"{attribute_value} = {n_Ai}")
                total += n_Ai
            print(f"Total = {total}\n")
```

# 6. Pelatihan P(B|Ai)

Pelatihan model melibatkan perhitungan probabilitas kondisional P(B|Ai), di mana B adalah kelas target yang akan diprediksi, dan Ai adalah fitur atau atribut yang akan digunakan sebagai dasar prediksi.

**P(B∣Ai)= P(Ai)/ P(Ai∣B)⋅P(B)**
​


```python
# Hasil Pelatihan P(B|Ai)

print("Hasil Pelatihan P(B|Ai)\n")

for attribute in attributes:
    if attribute != target_class:
        print("=" * 50)
        print(f"Hasil Pelatihan P(B|Ai) dari atribut \"{attribute}\"")
        print("=" * 50)
        for baris, target in enumerate(target_class_values):
            print(f"{attribute}")
            print(f"{target_class} \"{target}\" => P(B|A{baris + 1})")
            attribute_class_value = df[attribute].unique().tolist()
            total = 0
            print(f"{attribute} :")
            for attribute_value in attribute_class_value:
                n_Ai = len(df[(df[attribute] == attribute_value) & (df[target_class] == target)])
                n_Ai_target = len(df[df[target_class] == target])
                p_B_Ai = n_Ai / n_Ai_target
                print(f"{attribute_value} = {p_B_Ai}")
                total += p_B_Ai
            print(f"Hasil = {total}")
            if(baris < len(target_class_values) - 1):
                print("-" * 50)
        print("\n")


hasil_pelatihan = {}

for target in target_class_values:
    hasil_pelatihan[target] = {}
    for attribute in attributes:
        if attribute != target_class:
            hasil_pelatihan[target][attribute] = {}
            attribute_class_value = df[attribute].unique().tolist()
            for attribute_value in attribute_class_value:
                n_Ai = len(df[(df[attribute] == attribute_value) & (df[target_class] == target)])
                n_Ai_target = len(df[df[target_class] == target])
                p_B_Ai = n_Ai / n_Ai_target
                hasil_pelatihan[target][attribute][attribute_value] = p_B_Ai

print(hasil_pelatihan)
```

# 7. Prepare Data Testing

Mempersiapkan data yang akan digunakan untuk menguji performa model setelah melalui tahap pelatihan


```python
# import data untuk di test
data_test_file = './data/DataTest.xlsx'

df_test = pd.read_excel(data_test_file)

# Menampilkan 10 data teratas
df_test.head(10)
```

# 8. Pengujian

Pengujian performa model yang dihasilkan berdasarkan data testing

Berikut beberapa tahapan :

- Perhitungan P(B|Ai) sesuai dengan dataset yang akan di test
- Menghitung Posterior 
- Penarikan Kesimpulan


```python
# Menghitung P(B|Ai)
probability = []

for baris, data in df_test.iterrows():
    probability.append({})
    for target in target_class_values:
        probability[baris][target] = {}
        for attribute in attributes:
            if attribute != target_class:
                probability[baris][target][attribute] = hasil_pelatihan[target][attribute][data[attribute]]

for baris, data in df_test.iterrows():
    print(f"Data ke-{baris + 1}")
    for index, target in enumerate(target_class_values):
        total_probability = 1
        for attribute in attributes:
            if attribute != target_class:
                total_probability *= probability[baris][target][attribute]
        print(f"P(B|A{index + 1}) = {total_probability}")
    print("\n")
        
```

```python
# Menghitung Posterior P(B|Ai) . P(Ai) untuk data test berdasarkan hasil pelatihan

predicts = []

for baris, row in df_test.iterrows():
    print("=" * 100)
    print(f"Prediksi data test ke {baris + 1}")
    print("-" * 100)
    max_posterior = 0
    predict = ""
    for index, target in enumerate(target_class_values):
        total_probability = 1
        for attribute in attributes:
            if attribute != target_class:
                total_probability *= hasil_pelatihan[target][attribute][row[attribute]]
        print(f"P(B|A{index + 1}) = {total_probability}")
        print(f"P(A{index + 1}) = {prior[f'P(A{index + 1})']}")
        # Menghitung posterior
        posterior = total_probability * prior[f'P(A{index + 1})']
        print(f"Posterior {target_class} \"{target}\" = {posterior}\n")
        # Menentukan yang terbesar
        if posterior > max_posterior:
            max_posterior = posterior
            predict = target
    predicts.append(predict)
    print(f"Kesimpulan : \"{predict}\"\n")

    
```

```python
# Hasil Prediksi
df_test[target_class] = predicts
df_test.head(10)
```
